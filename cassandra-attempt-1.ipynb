```python
train_df=train_df.drop(columns=['UID','source_name','destination_name','trip_creation_time'])
```

```python
test_df=test_df.drop(columns=['UID','source_name','destination_name','trip_creation_time'])
```

```python
# remove the null values
Cat_value=[] 
Con_value = []
for i in train_df.columns:
  if train_df[i].dtype=='object':
    Cat_value.append(i)
  else :
    Con_value.append(i)
print(f'Columns which have Categorical Value is {Cat_value}')
print(f'Columns which have Continuos Value is {Con_value}')
```

```python
Con_value=['start_scan_to_end_scan', 'is_cutoff', 'cutoff_factor', 'actual_distance_to_destination', 'osrm_time', 'osrm_distance', 'factor', 'segment_actual_time', 'segment_osrm_time', 'segment_osrm_distance', 'segment_factor']
Cat_value=['route_schedule_uuid', 'route_type', 'trip_uuid', 'source_center', 'destination_center', 'od_start_time', 'od_end_time', 'cutoff_timestamp']
```

```python
for j in range(np.size(Cat_value)):
  train_df[Cat_value[j]].fillna(train_df[Cat_value[j]].mode()[0],inplace=True)
```

```python
for i in range(np.size(Con_value)):
  train_df[Con_value[i]].fillna(train_df[Con_value[i]].mean(),inplace=True)
```

```python
train_df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
for i in range(np.size(Cat_value)):
    
#     train_df[Cat_value[i]]=train_df[Cat_value[i]].values.reshape(-1,1)
    train_df[Cat_value[i]]= encoder.fit_transform(train_df[Cat_value[i]])
```

```python
# Now all the null values are removed
# Now Encoding Part
# Import label encoder
from sklearn import preprocessing
# label_encoder object knows how to understand word labels
label_encoder = preprocessing.LabelEncoder()
# Encode labels in column 'i'
for i in range(np.size(Cat_value)):
  test_df[Cat_value[i]]= label_encoder.fit_transform(test_df[Cat_value[i]])
```

```python
x_train = train_df.loc[:, train_df.columns != 'actual_time']
y_train = train_df.iloc[:,-1]
```

```python
x_test = test_df
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.ensemble import RandomForestRegressor
```

```python
# preprocessor=ColumnTransformer(transformers=[('num',StandardScaler(),Con_value),('cat',OneHotEncoder(handle_unknown='ignore'),Cat_value)])
# x_train_pre=preprocessor.fit_transform(x_train)
# scaler=StandardScaler(with_mean=False)
# x_train_scaled=scaler.fit_transform(x_train_pre)
model=RandomForestRegressor(n_estimators=100,random_state=42)
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
prdeicted_data = test_df1.copy()
prdeicted_data['actual_time'] = y_pred
```
